```python
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
all_points1 = []
all_points2 = []

with open('data/full_traj_obstacle.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        x1, y1 = float(row[4]), float(row[5])
        x2, y2 = float(row[7]), float(row[8])
        all_points1.append([x1, y1])
        all_points2.append([x2, y2])

num_trajectories = 1000
points_per_trajectory = 100

expert_data1 = [
    all_points1[i * points_per_trajectory:(i + 1) * points_per_trajectory]
    for i in range(num_trajectories)
]
first_trajectory1 = expert_data1[0]
x1 = [point[0] for point in first_trajectory1]
y1 = [point[1] for point in first_trajectory1]

expert_data2 = [
    all_points2[i * points_per_trajectory:(i + 1) * points_per_trajectory]
    for i in range(num_trajectories)
]
first_trajectory2 = expert_data2[0]
x2 = [point[0] for point in first_trajectory2]
y2 = [point[1] for point in first_trajectory2]

expert_data1 = np.array(expert_data1)
expert_data2 = np.array(expert_data2)

# Compute mean and standard deviation
combined_data = np.concatenate((expert_data1, expert_data2), axis=0)
mean = np.mean(combined_data, axis=(0,1))
std = np.std(combined_data, axis=(0,1))

# Normalize data
# expert_data1 = (expert_data1 - mean) / std
# expert_data2 = (expert_data2 - mean) / std
```

```python
# Get the first trajectory from expert_data1
first_trajectory1 = expert_data1[ind]
first_trajectory2 = expert_data2[ind]

# Define the amount of shift to be added
shift_magnitude = 0.5

# Generate 100 shifted trajectories
expert2_manual = np.array([
    first_trajectory2 + [np.random.uniform(-shift_magnitude, shift_magnitude), np.random.uniform(-shift_magnitude, 0)]
    for _ in range(1000)
])
expert1_manual = np.array([
    first_trajectory1 + [np.random.uniform(-shift_magnitude, shift_magnitude), np.random.uniform(-shift_magnitude, 0)]
    for _ in range(1000)
])


plt.figure(figsize=(20, 8))
for traj in expert1_manual[:]:  # Plot a few expert trajectories
    first_trajectory = traj
    x = [point[0] for point in first_trajectory]
    y = [point[1] for point in first_trajectory]
    plt.plot(x, y, 'b--')
for traj in expert2_manual[:]:  # Plot a few expert trajectories
    first_trajectory = traj
    x = [point[0] for point in first_trajectory]
    y = [point[1] for point in first_trajectory]
    plt.plot(x, y, 'g--')

plt.show()
```
