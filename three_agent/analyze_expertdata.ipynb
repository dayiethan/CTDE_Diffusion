```python
import numpy as np
import matplotlib.pyplot as plt

# Replace these with your actual arrays of shape (100, 100, 2)
# For demonstration purposes, we're creating dummy arrays here:
arr1 = np.random.randn(100, 100, 2)
arr2 = np.random.randn(100, 100, 2)
arr3 = np.random.randn(100, 100, 2)

def plot_all(arr1, arr2, arr3):
    """
    Plots the first trajectory (index 0) from each of three (100,100,2) arrays.
    """
    
    plt.figure(figsize=(8, 6))
    plt.xlim(-3, 5)
    plt.ylim(-3, 3)
    for i in range(len(arr1)):
        t1 = arr1[i]
        t2 = arr2[i]
        t3 = arr3[i]
        plt.scatter(t1[0, 0], t1[0, 1], label='Array 1 Traj 0', color='blue')
        # plt.plot(t1[:, 0], t1[:, 1], label='Array 1 Traj 0', color='blue')
        plt.scatter(t2[0, 0], t2[0, 1], label='Array 2 Traj 0', color='orange')
        # plt.plot(t2[:, 0], t2[:, 1], label='Array 2 Traj 0', color='orange')
        plt.scatter(t3[0, 0], t3[0, 1], label='Array 3 Traj 0', color='green')
        # plt.plot(t3[:, 0], t3[:, 1], label='Array 3 Traj 0', color='green')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('First Trajectory from Each Array')
    plt.grid(True)
    plt.show()

# Example usage
plot_all(expert_data_1, expert_data_2, expert_data_3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Palatino Linotype'] + plt.rcParams['font.serif']

def plot_expert(arr1, arr2, arr3):
    """
    Plots the paths of three agents.
    - Earlier trajectories muted.
    - Final trio emphasized with bold, thicker lines.
    """
    plt.figure(figsize=(8, 6))
    ax = plt.gca()

    # Define colors (normal vs muted)
    colors = {
        "Agent 1": ("blue", "lightblue"),
        "Agent 2": ("orange", "navajowhite"),
        "Agent 3": ("green", "lightgreen"),
    }

    highlight = 10

    for i in range(len(arr1)):
        if i != highlight:
            # Earlier: muted thin lines
            plt.plot(arr1[i][:, 0], arr1[i][:, 1], color=colors["Agent 1"][1], alpha=0.4, linewidth=1)
            plt.plot(arr2[i][:, 0], arr2[i][:, 1], color=colors["Agent 2"][1], alpha=0.4, linewidth=1)
            plt.plot(arr3[i][:, 0], arr3[i][:, 1], color=colors["Agent 3"][1], alpha=0.4, linewidth=1)

    plt.plot(arr1[highlight][:, 0], arr1[highlight][:, 1], color=colors["Agent 1"][0], linewidth=3, label="Agent 1")
    plt.plot(arr2[highlight][:, 0], arr2[highlight][:, 1], color=colors["Agent 2"][0], linewidth=3, label="Agent 2")
    plt.plot(arr3[highlight][:, 0], arr3[highlight][:, 1], color=colors["Agent 3"][0], linewidth=3, label="Agent 3")

    # Labels and title with larger fonts
    plt.xlabel("X-Position", fontsize=16)
    plt.ylabel("Y-Position", fontsize=16)
    plt.title("Three-Agent Road Crossing", fontsize=18)

    # Legend formatting
    plt.legend(frameon=False, fontsize=14)

    # Remove grid and extra spines
    plt.grid(False)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.show()

# Example call
plot_expert(expert_data_1, expert_data_2, expert_data_3)
```
