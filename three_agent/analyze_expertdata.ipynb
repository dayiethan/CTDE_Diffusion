```python
expert_data_1 = np.load('data/expert_data1_100_traj.npy')
expert_data_2 = np.load('data/expert_data2_100_traj.npy')
expert_data_3 = np.load('data/expert_data3_100_traj.npy')
print(expert_data_1.shape, expert_data_2.shape, expert_data_3.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace these with your actual arrays of shape (100, 100, 2)
# For demonstration purposes, we're creating dummy arrays here:
arr1 = np.random.randn(100, 100, 2)
arr2 = np.random.randn(100, 100, 2)
arr3 = np.random.randn(100, 100, 2)

def plot_three_first(arr1, arr2, arr3, i):
    """
    Plots the first trajectory (index 0) from each of three (100,100,2) arrays.
    """
    t1 = arr1[i]  # shape (100, 2)
    t2 = arr2[i]
    t3 = arr3[i]

    plt.figure(figsize=(8, 6))
    plt.plot(t1[:, 0], t1[:, 1], label='Array 1 Traj 0')
    plt.plot(t2[:, 0], t2[:, 1], label='Array 2 Traj 0')
    plt.plot(t3[:, 0], t3[:, 1], label='Array 3 Traj 0')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('First Trajectory from Each Array')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_three_first(expert_data_1, expert_data_2, expert_data_3, i = 6)
```
